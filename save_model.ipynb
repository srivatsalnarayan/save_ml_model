```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/save_model/homeprices.csv")
df.head()
```

```python
model = linear_model.LinearRegression()
model.fit(df[['area']],df.price)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[5000]])
```

Save Model
using pickle

```python
import pickle
```

```python
with open('model_pickle','wb') as file:
    pickle.dump(model,file)
```

```python
with open('/content/drive/MyDrive/Colab Notebooks/save_model/model_pickle','rb') as file:
    mp = pickle.load(file)
```

```python
mp.predict([[5000]])
```

Save Trained Model Using joblib
